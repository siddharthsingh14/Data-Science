# Job Location
Given File 'amazonjobsdataset.csv'<br>
It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.
##### Problem Statement :
Find number of job openings in Bangalore,IN and in Seattle,US?
Print the Number of Job opening in Bangalore and Seattle as Integer value.
#### Output Format :
CountBangalore CountSeattle

```python
## Open and read data file as specified in the question
## Print the required output in given format
import csv

with open('amazon_jobs_dataset.csv',encoding='utf-8') as file_obj :
    file_list = list(csv.DictReader(file_obj, skipinitialspace = True))
    
count = 0
countb = 0
counts = 0
for row in file_list:
    if(row['location'].strip() in ["US, WA, Seattle", "IN, KA, Bangalore"]):
        count += 1
    if(row['location'].strip() == "US, WA, Seattle"):
        counts += 1
    if(row['location'].strip() in ["IN, KA, Bangalore"]):
        countb += 1

print(countb,counts)
```

# Job Month 2018
Given File 'amazonjobsdataset.csv'<br>
It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.
#### Problem Statement :
Find the month having most job openings in Year 2018 ?<br>
Print the month (Month Name i.e January, February, March) and Number of job opening as Integer Value
#### Output Format :
MonthName Count

```python
import csv

with open('amazon_jobs_dataset.csv',encoding='utf-8') as file_obj :
    file_list = list(csv.DictReader(file_obj, skipinitialspace = True))

dic = {}
count = 0
for row in file_list:
    post_date = row['Posting_date'].strip().split()
    if post_date[2] == '2018':
        if post_date[0] in dic:
            dic[post_date[0]] += 1
        else:
            dic[post_date[0]] = 1
        
MAX=-11111111111
MONTH = ''
for i in dic:
    if dic[i] > MAX:
        MAX = dic[i]
        MONTH = i
    
    else:
        continue
        
else:
    print(MONTH,MAX)
```

# Job Degree
Given File 'amazonjobsdataset.csv'<br>
It is a dataset including information on amazon job opening around the world from June 2011 to March 2018. This dataset is collected using Selenium and BeautifulSoup by scraping all of the jobs for Amazon job site.
#### Problem Statement :
Find the number of job openings are present if applicant have Bachelor degree?<br>
Print the count as Integer value<br>
Note : Here we will use the BASIC QUALIFICATIONS feature to find out whether bachelor degree for Job is required or not. <br>
Keywords that can be used are 'Bachelor', 'BS' and 'BA'.
#### Output Format :
Count

```python
import csv

with open('amazon_jobs_dataset.csv',encoding='utf-8') as file_obj :
    file_list = list(csv.DictReader(file_obj, skipinitialspace = True))
    
count = 0
for row in file_list:
    quali = (row['BASIC QUALIFICATIONS'].strip())
    if quali.find('BS') != -1:
        count += 1
    elif quali.find('Bachelor') != -1:
        count += 1
    elif quali.find('BA') != -1:
        count += 1
print(count)
```
